```python
from tkinter import *
from tkinter import messagebox
from tkinter import filedialog
import PIL.Image, PIL.ImageTk
import cv2
from threading import Thread

# TODO: get the source for this code
class videoGUI:

    def __init__(self, window, window_title):

        self.window = window
        self.window.title(window_title)

        left_frame = Frame(self.window)
        left_frame.pack(side=TOP, pady=5)

        right_frame = Frame(self.window)
        right_frame.pack(side=BOTTOM, pady=5)

        bottom_frame = Frame(self.window)
        bottom_frame.pack(side=BOTTOM, pady=5)

        self.pause = False   # Parameter that controls pause button

        self.left_canvas = Canvas(left_frame)
        self.left_canvas.pack()

        self.right_canvas = Canvas(right_frame)
        self.right_canvas.pack()

        # Select Button
        self.btn_select=Button(bottom_frame, text="Select video file", width=15, command=self.open_file)
        self.btn_select.grid(row=0, column=0)

        # Play Button
        self.btn_play=Button(bottom_frame, text="Play", width=15, command=self.play_video)
        self.btn_play.grid(row=0, column=1)

        # Pause Button
        self.btn_pause=Button(bottom_frame, text="Pause", width=15, command=self.pause_video)
        self.btn_pause.grid(row=0, column=2)

        self.delay = 5   # ms
        
#         self.webcam_thread = Thread(target=self.play_webcam,args=(self))
#         self.webcam_thread.start()

        self.window.mainloop()


    def open_file(self):

        self.pause = False

        self.filename = filedialog.askopenfilename(title="Select file", filetypes=(("MP4 files", "*.mp4"),
                                                                                         ("WMV files", "*.wmv"), ("AVI files", "*.avi")))
        print(self.filename)

        # Open the video file
        self.cap = cv2.VideoCapture(self.filename)
        self.webcam = cv2.VideoCapture(0)

        self.cap = cv2.resize(self.cap,(640,360))

        self.width = int(self.cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        self.height = int(self.cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

        self.width_webcam = self.webcam.get(cv2.CAP_PROP_FRAME_WIDTH)
        self.height_webcam = self.webcam.get(cv2.CAP_PROP_FRAME_HEIGHT)

        self.left_canvas.config(width = 640, height = 360)
        self.right_canvas.config(width = 640, height = 360)


    def get_frame(self):   # get only one self.frame

        try:

            if self.cap.isOpened():
                self.ret, self.frame = self.cap.read()
                self.frame = cv2.resize(self.frame,(640,360))
                self.webcam_ret, self.webcam_frame = self.webcam.read()
                self.webcam_frame = cv2.resize(self.webcam_frame,(640,320))

                return (self.ret, cv2.cvtColor(self.frame, cv2.COLOR_BGR2RGB),self.webcam_ret, cv2.cvtColor(self.webcam_frame, cv2.COLOR_BGR2RGB))

        except:
            messagebox.showerror(title='Video file not found', message='Please select a video file.')


    def play_video(self):

        # Get a self.frame from the video source, and go to the next self.frame automatically
        self.ret, self.frame,self.ret2, self.frame2  = self.get_frame()

        if self.ret:
            self.photo = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.frame))
            self.photo2 = PIL.ImageTk.PhotoImage(image = PIL.Image.fromarray(self.frame2))
            self.left_canvas.create_image(0, 0, image = self.photo, anchor = NW)
            self.right_canvas.create_image(0, 0, image = self.photo2, anchor = NW)

        if not self.pause:
            self.window.after(self.delay, self.play_video)
    
#     def get_webcam_frame(self):
#         try:
#             if self.webcam.isOpened():
#                 flag,frame = self.cap.read()
#                 frame = cv2.resize(frame,(640,320))
#                 return flag, cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
#         except:
#             messagebox.showerror(title='Webcam Not On',message='Restart bc the webcam is not on')
    
#     def play_webcam(self):
#         webcam_flag = False # init value so while loop runs w out get_webcam_frame doing anything
#         while True:
#             webcam_flag,webcame_frame = self.get_webcam_frame()
#             # webcam_flag = True if we could find a webcam frame
#             if webcam_flag == True:
#                 # show the webcam
#                 self.right_canvas.create_image(0, 0, image = self.photo2, anchor = NW)
#             if not self.pause:
#                 self.window.after(self.delay, self.play_webcam)
    def pause_video(self):
        self.pause = True


    # Release the video source when the object is destroyed
    def __del__(self):
        if self.cap.isOpened():
            self.cap.release()
        self.webcam_thread.join() # wait for thread to end

##### End Class #####


# Create a window and pass it to videoGUI Class
videoGUI(Tk(), "Happy CV")
```
